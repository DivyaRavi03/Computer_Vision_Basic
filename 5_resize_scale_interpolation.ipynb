## Resize, Scaling, Interpolation 

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("../images/test_img.jpg")
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # must change the image to rgb or else further functions will be handled only with bgr colors
```

### Resize 

```python
image.shape # to know the image size (height, width, channels)
```

```python
height, width = 400, 400 # define height and width that needs the image to be resized for 

resize_img = cv2.resize(rgb_image, (width, height))

plt.imshow(resize_img)
```

## Image Scaling 

Unlike resize, we need to give scale percentage, multiply or divide it with respective of upscale and downscale and then again use the resize methos to resize image but with new calculated scaled values 

```python
scale_per = 50
s_width = int(rgb_image.shape[1]*scale_per/100)
s_height = int(rgb_image.shape[0]*scale_per/100)

scaled_img = cv2.resize(rgb_image, (s_width, s_height), interpolation=cv2.INTER_AREA) # samples are lost when upscaling, not in downscaling

print(scaled_img.shape)
plt.imshow(scaled_img)
```

### Interpolation

Interpolation - estimate the values to unknown pixels with the known data. 

```python
resize_width = 600
resize_height = 600 

# interpolation methods 
methods = {
    'Nearest': cv2.INTER_NEAREST, 
    'Linear': cv2.INTER_LINEAR, 
    'Cubic': cv2.INTER_CUBIC, 
    'Area': cv2.INTER_AREA, 
    'Lanczos4': cv2.INTER_LANCZOS4
}
ip_image = {}
for name, method in methods.items():
    ip_image[name] = cv2.resize(rgb_image, (resize_width,resize_height), interpolation=method)

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

for i,(name, ip_image) in enumerate(ip_image.items(), start = 2):
    plt.subplot(2,3,i)
    plt.imshow(ip_image)
    plt.title(name)
    plt.axis('off')

plt.tight_layout()
plt.show()

```
