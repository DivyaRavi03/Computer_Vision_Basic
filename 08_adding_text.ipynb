# Adding text on images

```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
height, width = 500, 700
sky_blue = (255, 127, 0)
canva = np.full((height, width, 3), sky_blue, dtype = np.uint8)
canva = cv2.cvtColor(canva, cv2.COLOR_BGR2RGB)

plt.imshow(canva)
```

```python
text = "Hello Opencv"
font = cv2.FONT_HERSHEY_COMPLEX

# coordinate for bototm left corner of the text 
org = (50, 100)
font_scale = 1.5 # default * with font_scale 
color = (255,255, 255)
thickness = 2 

cv2.putText(canva, text, org, font, font_scale, color, thickness, cv2.LINE_AA) # line for rendering text  
plt.imshow(canva)
```

## Different fonts 

```python
canva = np.full((height, width, 3), sky_blue, dtype = np.uint8)
canva = cv2.cvtColor(canva, cv2.COLOR_BGR2RGB)

fonts = [
    cv2.FONT_HERSHEY_SIMPLEX,
    cv2.FONT_HERSHEY_COMPLEX, 
    cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
    cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 
    cv2.FONT_HERSHEY_COMPLEX_SMALL
]
# making all the text to be in same x, but different in y axes 

y_offset = 50 
for i, font in enumerate(fonts):
    text = f"Font {i+1}"
    # cv2.putText(image, text, starting point (x, y), font, scale, color, thickness, line to render)
    cv2.putText(canva, text, (50, y_offset), font, 1, (0,0,255), 2, cv2.LINE_AA)
    y_offset +=50

plt.imshow(canva)
```
