# Blurring Images

3 types of blurring: 
1. Average Blurring - takes the average of neighbor pixels to get the pixel for output image ( Smoothening and Noise)
2. Gaussian Blurring- takes weighted average around the center of image ( Smoothen, more quality and natural image)
3. Median Blurring - takes median of surrounding pixels to get the pixel for output image (Better for noise reduction)

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/test_img.jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Average Blur

Bigger the kernel size, more blurriness effect will be 

```python
kernel_size = 5
average_blur = cv2.blur(image_gray,(kernel_size, kernel_size))
plt.imshow(average_blur, cmap="gray")
```

## Gaussian Blur

sigma x = 0 - standard deviation of the gaussian in x direction; y = 0 in y direction (higher sigma, the smoother blur ); kernel size must be in odd 

```python
kernel_size = 31
gaussian_blur = cv2.GaussianBlur(image_gray, (kernel_size, kernel_size),0 )
plt.imshow(gaussian_blur, cmap="gray")
```

## Median Blur

```python
kernel_size = 5
median_blur = cv2.medianBlur(image_gray, kernel_size )
plt.imshow(median_blur, cmap="gray")
```

```python
plt.figure(figsize=(14,12))
plt.subplot(2,2,1)
plt.title("Grayscale original")
plt.imshow(image_gray, cmap="gray")
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Average Blur")
plt.imshow(average_blur, cmap="gray")
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Gaussian Blur")
plt.imshow(gaussian_blur, cmap="gray")
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Median blur")
plt.imshow(median_blur, cmap="gray")
plt.axis("off")
```
