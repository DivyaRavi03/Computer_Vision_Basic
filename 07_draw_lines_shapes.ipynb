## Draw lines and shapes

```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np  
```

## Create an empty canvas

```python
height, width = 500, 700
blue = (255, 127, 0)
canva = np.full((height, width, 3), blue, dtype = np.uint8)

plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

## Draw a line

cv2.line(image, starting, ending, color of line, thickness)

```python
cv2.line(canva, (50,50), (650,450), (255, 255, 255), 5)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

## Draw a rectangle

cv2.rectangle(image, (x,y), (x,y), color, thickness)

```python
cv2.rectangle(canva, (100,400), (300,100), (0,255,255), 3)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

```python
cv2.rectangle(canva, (100,100), (300,300), (0,255,255), 5)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

## Draw a circle

cv2.circle(image, (centre point), radius, color, thickness)

```python
cv2.circle(canva, (500, 150), 100, (0,0,255), 5)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

## Draw a ellipse

cv2.ellipse(image, center point, length of majpr axes, minor axes, angle, start angle, end angle, color, thickness)

```python
cv2.ellipse(canva, (350, 350), (100,50), 45, 0, 360, (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```

## Draw a polygon

xn, yn, where n>2

```python
points = np.array([[150, 400], [300,400], [255, 300]], np.int32)
reshaped = points.reshape((-1, 1, 2))

print(reshaped)

cv2.polylines(canva, [reshaped], True, (255, 0, 255), 3)
plt.imshow(cv2.cvtColor(canva, cv2.COLOR_BGR2RGB))
```
