# Exploring different color spaces that are managed by OpenCV 

```python
import cv2
```

```python
bgr_image = cv2.imread("../images/test_img.jpg") # read the image into numoy array from the given path 
```

```python
# Convert to grayscale 
gray_scale = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)
cv2.imwrite("../images/gray_img.jpg", gray_scale)
```

```python
# Convert to HSV 
hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
cv2.imwrite("../images/hsv_img.jpg", hsv_image )
```

```python
# Convert to LAB
lab_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2LAB)
cv2.imwrite("../images/lab_img.jpg", lab_image)
```

```python
# Convert to RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB) # because opencv saves and reads in default as bgr
cv2.imwrite("../images/rgb_img.jpg", rgb_image)
```

```python
bgr_image_2 = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2BGR)
cv2.imwrite("../images/to_bgr_img.jpg", bgr_image_2)
```

if input is RGB, opencv will convert to BGR.
cv2.imshow() will consume bgr but displays in the form of rgb because humans perceive that color, imwrite()
