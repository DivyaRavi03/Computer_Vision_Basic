# Image Filters 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/filter.jpg", cv2.IMREAD_GRAYSCALE) # converting directly to grayscale 
plt.imshow(image, cmap="gray")
```

from this filter we want to detect onky the horizontal edges, with higher intensity in the middle it works well by neturalizing 

```python
# custom kernel 
# horizontal_kernel = np.array([[-1, -1, -1], [0,0,0], [1,1,1]]) # intensity are not affected so much, because more intensity it is better to extract information [-10, 10, 10]
horizontal_kernel = np.array([[-1, -1, -1], [-10,-10,-10], [1,1,1]])
# to apply cnn on filter over the image 
horizontal_image = cv2.filter2D(image, -1, horizontal_kernel)

plt.imshow(horizontal_image, cmap="gray")
```

```python
x = -10 
# vertical kernel 
vertical_kernel = np.array([[-1, x, 1], [-1, x, 1], [-1, x, 1]]) # must be enclosed within -1, 1 
vertical_image = cv2.filter2D(image, -1, vertical_kernel)
plt.imshow(vertical_image, cmap="gray")
```

```python
diagonal_kernel = np.array([[x,-1, -1], [1, x, -1], [1,1,x]]) # all diagonal must be x, before x must be 1, after x must be -1 
diagonal_image = cv2.filter2D(image, -1, diagonal_kernel)
plt.imshow(diagonal_image, cmap="gray")
```
