# Edge Detection

3 types: 1. Sobel Technique
         2. Laplacian Technique
         3. Canny Edge Detector

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/test_img.jpg", cv2.IMREAD_GRAYSCALE)
```

## Sobel Edge Detector 

```python
# X direction 
sobel_x = cv2.Sobel(image, cv2.CV_64F,1,0, ksize = 3) #output depth of values - to calculate sobel in x direction (1); y = 0

# Y direction 
sobel_y = cv2.Sobel(image, cv2.CV_64F,1,0, ksize = 3) # to calculate sobel in y direction (1); x = 0

# magnitude matrix 
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
```

## Laplacian Edge Detector

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize = 3) # it need only positive values
laplacian_abs = cv2.convertScaleAbs(laplacian)
```

## Canny Edge Detector

defines edges well and also reduces the noise well 

```python
canny = cv2.Canny(image, 180, 200) # canny(image, lower_threshold, higher_threshold)
```

## Displaying all the edges 

```python
title = ["Original","Sobel","Laplacian_all","Laplacian Abs","Canny"]
images = [image, sobel_combined, laplacian, laplacian_abs, canny]

plt.figure(figsize=(20,16))

for i in range(len(title)):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(title[i])
    plt.axis("off")

plt.show()
```
