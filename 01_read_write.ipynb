# Read, Display, Write images


Opencv - to process images and videos, detect objects, faces and even handwriting
       - Comprehensive Image Processing: process, analyze, manipulate the images
       - Cross-Platform, Open source
       - Intergrated with DL frameworks

```python
import cv2
```

```python
# Read the image
image = cv2.imread("../images/test_img.jpg") # load the image to matrix of pixels that returns into numpy array 
```

```python
# check if the image is loaded 
if image is None:
    print("Error")
else:
    print("Image loaded successfully ")
```

```python
# Display image 
cv2.imshow("Image", image) # displays the image with title 
cv2.waitKey(0) # pauses window for a while and closes when a key is pressed 
cv2.destroyAllWindows() # closes all the cv windows
```

```python
# Write image 
image_new = image + 30 # adding + 30 to all pixels 
cv2.imwrite("../images/wrote_img.jpg", image_new)
```
