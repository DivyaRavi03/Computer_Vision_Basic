## Flip, Rotate, Crop

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/test_img.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

## Flip image (Mirroring)

```python
# 1 --> horizontal; 0 -> vertical; -1 -> both axes
flip_horizontal = cv2.cvtColor((cv2.flip(image, 1)), cv2.COLOR_BGR2RGB)
flip_vertical = cv2.cvtColor((cv2.flip(image, 0)), cv2.COLOR_BGR2RGB)
flip_both = cv2.cvtColor((cv2.flip(image, -1)), cv2.COLOR_BGR2RGB)

# flip_horizontal = cv2.flip(image_rgb,1)
# flip_vertical = cv2.flip(image_rgb,0)
# flip_both = cv2.flip(image_rgb,-1)

titles = ["Original","Horizontal","Vertical", "Both"]
images = [image_rgb, flip_horizontal, flip_vertical, flip_both ]

plt.figure(figsize=(12,8))
for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout
plt.show()
```

## Rotation (Aligning)

```python
rotate_90_clock = cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE)
rotate_90_counter_clock = cv2.rotate(image_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180_clock = cv2.rotate(image_rgb, cv2.ROTATE_180)

titles = ["Original","90 d - clockwise","90 - d - counter clock", "180 d"]
images = [image_rgb, rotate_90_clock, rotate_90_counter_clock, rotate_180_clock ]

plt.figure(figsize=(12,8))
for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
```

## Crop an image (Specific part of an image )

```python
print(image_rgb.shape) # height, width, channel 
```

```python
x_start, y_start = 400, 300
x_end, y_end = 800, 900

crop_img = image_rgb[y_start: y_end, x_start:x_end]

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(crop_img)
plt.title("Cropped")
plt.axis("off")

plt.tight_layout()
plt.show()
```
