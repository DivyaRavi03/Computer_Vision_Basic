# Working with video files

```python
import cv2
```

```python
# Load video
video = cv2.VideoCapture("../images/video_1.mp4") # storing the loaded video in video object, helps to retrieve frames one by one 
```

```python
# Display video, Because you keep reading frames until the video ends
while True:
    ret, frame = video.read() # after reading, 1. ret reuturns bool value of frame 2. frame is captured, it returns 
    # if the file is wrong or has some issue or the file ended ret returns false, else true 
    if not ret: 
        break
    cv2.imshow("Video Frames", frame) # displaying title and numpy array stored as frame
    # break out of while loop
    if cv2.waitKey(1) & 0xFF == ord('q'): # waitKey(1) - waits for 1ms b/w frames
        break 

video.release # release video object 
cv2.destroyAllWindows() # destroy all windows created, because it has created a single video but multiple times of frames

```

```python
# Write a video
video = cv2.VideoCapture("../images/video_1.mp4")
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH)) # returns width of object
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT)) 

video_output = cv2.VideoWriter("../images/wrote_video.mp4", fourcc, 30, (width, height))

while True:
    ret, frame = video.read()
    if not ret: 
        break 
    video_output.write(frame)

video_output.release()
video.release()
cv2.destroyAllWindows()
```
